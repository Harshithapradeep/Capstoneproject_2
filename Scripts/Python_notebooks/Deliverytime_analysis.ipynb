```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from CSV file
file_path = r'C:\Users\hbhat\OneDrive\Desktop\VScode\Dummy files\Sales.csv'  
data = pd.read_csv(file_path)

# Ensure 'delivery_date' and 'order_date' columns are converted to datetime format
data['Delivery Date'] = pd.to_datetime(data['Delivery Date'], errors='coerce')
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')

# Drop rows with missing or invalid dates in either column
data=data.dropna(subset=['Delivery Date', 'Order Date'])

# Calculate the difference in days between 'delivery_date' and 'order_date'
data.loc[:, 'delivery_time_days']= (data['Delivery Date'] - data['Order Date']).dt.days

# Plotting the histogram for the difference in days
plt.figure(figsize=(8,4))
sns.histplot(data['delivery_time_days'],bins=18, kde=False)


plt.title('Average Delivery Time Analysis')
plt.xlabel('Delivery Time (Days)')
plt.ylabel('Frequency')
plt.tight_layout()

# Show the plot
plt.show()

# Min and max delivery days
Max_delivery_time=data.loc[:, 'delivery_time_days'].max()   
Min_delivery_time=data.loc[:, 'delivery_time_days'].min() 
print(Max_delivery_time)   
print(Min_delivery_time)

# Average Delivery time analysis -Sum of delivery days/count delivery time values
total=data.loc[:, 'delivery_time_days'].sum()    
count =data.loc[:, 'delivery_time_days'].count()
print(total/count)
```
